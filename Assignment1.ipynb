# Tasks of a Data Analyst
- Collecting Data
    - Gather data from databases, spreadsheets, APIs, or internal tools
    - Work with raw data from systems (sales, customers, logs, etc.)
    - Example: Pulling user activity data from a company database
- Cleaning & Preparing Data
    - Remove duplicates, fix errors, handle missing values
    - Convert messy data into usable format
    - This is where a lot of time goes (honestly like 60–70%)
- Data Analysis
    - Look for patterns, trends, and insights
    - Use statistics to understand what’s happening
    - Example: “Why did sales drop last month?”
- Data Visualization
    - Turn numbers into charts, dashboards, and graphs
    - Make data easy to understand for non-technical people
    - Tools often used: Tableau and Microsoft Power BI
- Reporting & Communication
    - Present findings to managers or teams
    - Explain insights in simple terms
    - Example: “Customer churn increased because of delayed delivery times”
- Working with Databases
    - Query databases using SQL
    - Extract specific data needed for analysis
    - Example: Writing queries to get last 6 months of transactions
- Supporting Decision-Making
    - Help teams make business decisions based on data
    - Provide recommendations
    - Example: “We should focus marketing on this age group”
                             

# Skills of a Data Analyst

- Technical Skills
    - Excel (Must-have)
        - Pivot tables, formulas, data cleaning
        - Still heavily used in companies
    - SQL (Very Important)
        - Query databases
        - Filter, join, and aggregate data
    - Programming
        - Python (most popular)
        - R (used for statistics)
            - Used for:
                - Data cleaning
                - Automation
                - Analysis
    - Data Visualization Tools
        - Tableau
        - Microsoft Power BI
    - Statistics Basics
        - Mean, median, standard deviation
        - Correlation, probability
        - Helps you interpret data correctly

- Analytical Skills
    - Critical thinking
    - Problem-solving
    - Attention to detail
        - You’re basically solving puzzles with data

- Soft Skills
    - Communication (very important)
    - Storytelling with data
    - Business understanding
        - You need to explain insights clearly to non-tech people

Simple Way to Think About It:
    - A Data Analyst:
        - Takes messy data → cleans it → analyzes it → explains what it means → helps decisions



# Steps of doing data analysis

- Understand the Problem (Business Question)
    - First, clarify what you’re trying to solve
    - Talk to stakeholders or manager
    - Example: 
        - “Why are customer sign-ups dropping this month?”

- Collect the Data
    - Gather data from:
    - Databases (using SQL)
    - Excel files
    - APIs or logs
    - Example:
        - Pull user signup data from last 6 months

- Clean the Data
    - Fix missing values
    - Remove duplicates
    - Correct wrong formats
    - This step is super important—bad data = wrong results

- Explore the Data (EDA – Exploratory Data Analysis)
    - Look at patterns, trends, outliers
    - Use basic stats and quick charts
    - Example:
        - Are sign-ups lower on certain days?
        - Any sudden drop after a specific date?

- Analyze the Data
    - Dig deeper using logic and statistics
    - Compare different variables
    - Example:
        - Compare sign-ups by region, device, or campaign
        - Identify what’s causing the drop

- Visualize the Data
    - Create charts, dashboards, and graphs
    - Make it easy for others to understand
    - Common tools:
        - Tableau
        - Microsoft Power BI

- Interpret Results
    - Turn numbers into insights
    - Answer the original question clearly
    - Example:
        - “Sign-ups dropped due to a bug in the mobile app after the last update”

- Present Findings
    - Share results with stakeholders
    - Use simple language (not too technical)
    - Could be:
        - Report
        - Dashboard
        - Presentation

- Make Recommendations
    - Suggest what should be done next
    - Example:
        - Fix mobile bug
        - Improve onboarding process

- Monitor & Update
    - Track results after changes
    - Update dashboards regularly
        - Data analysis is continuous, not one-time


# Easy Way to Remember

- Ask → Get Data → Clean → Analyze → Explain → Recommend


# Real-Life Flow Example

- Problem: Sales dropped
    - Ask why
    - Get sales data
    - Clean it
    - Analyze trends
    - Find issue (e.g., pricing change)
    - Show chart
    - Explain to team
    - Suggest fix
