```python
data = [
    [1, 'Taheera', 70, 86], [2, 'Aelita', 56, 90], [3, 'Khulan', 87, 90], [4, 'Batgerel', 71, 79],[5, 'Manisha', 88, 76]]
df = pd.DataFrame(data, columns=['Rollno', 'Name', 'Maths', 'English'])
```

```python
df
```

```python
df_long = df.melt(id_vars=['Rollno','Name'],value_vars=['Maths','English'],var_name='Subject',value_name='Marks')
df_long
```

```python
wide_df = df_long.pivot(index=['Rollno','Name'],values='Marks').reset_index()
wide_df
```

```python
wide_df = df_long.pivot(index=['Rollno','Name'],columns='Subject',values='Marks').reset_index()
wide_df
```

```python
wide_df = df_long.pivot(index=['Rollno','Name'], columns='Subject',
             values='Marks').reset_index(drop=False).rename_axis(None, axis=1)
wide_df
```

```python
wide_df2 = df_long.pivot_table(index=['Rollno','Name'], columns='Subject',
                               values='Marks', aggfunc='sum').reset_index(drop=False).rename_axis(None, axis=1)
wide_df2
```
