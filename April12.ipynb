```python
mtcars = data('mtcars')
mtcars = mtcars.astype({'cyl':'category', 'vs':'category', 'carb':'category', 'am':'category', 'gear': 'category'})
```

```python
# encode - map
mtcars['am'] = mtcars['am'].map({0: 'auto', 1: 'manual'})
mtcars.head()
#mtcars['am'] = mtcars['am'].replace({0: 'auto', 1: 'manual'})
```

```python
mtcars['gear'] = mtcars['gear'].map({3: 'Gear3', 4: 'gear4', 5: 'gear5'})
mtcars.head()
```

```python
bins = [0, 17, 29, mtcars['mpg'].max()]
labels = ['Low', 'Medium', 'High']

# Create the new categorical column
mtcars['mpg_level'] = pd.cut(mtcars['mpg'], bins=bins, labels=labels, include_lowest=True)
```
